# Import Libraries & Modules

```python
import torch
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader, TensorDataset

import matplotlib.pyplot as plt
import numpy as np

from torchmetrics import Accuracy

torch.manual_seed(0)
```

# Load Dataset and read images 

```python
train_transform = transforms.Compose([transforms.Resize(size = (224, 224)),
                                      transforms.RandomHorizontalFlip(),
                                      transforms.ToTensor(),
                                      transforms.Normalize(mean=[0.4954, 0.4954, 0.4955],
                                                            std=[0.2801, 0.2801, 0.2801])])

test_transform = transforms.Compose([transforms.Resize(size = (224, 224)),
                                     transforms.ToTensor(),
                                     transforms.Normalize(mean=[0.4954, 0.4954, 0.4955],
                                                          std=[0.2801, 0.2801, 0.2801])])
```

```python
train_data = torchvision.datasets.ImageFolder('/content/drive/MyDrive/Datasets/Covid19-dataset/train',
                                             transform=train_transform)

test_data = torchvision.datasets.ImageFolder('/content/drive/MyDrive/Datasets/Covid19-dataset/test',
                                            transform=test_transform)
```

## DataLoader

```python
batch_size = 6

train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_data, batch_size=11, shuffle=False)

print('Number of training batches', len(train_loader))
print('Number of test batches', len(test_loader))
```

# Show Images

```python
class_names = train_data.classes

def show_images(images, labels, preds):
    plt.figure(figsize=(16, 9))
    for i, image in enumerate(images):
        plt.subplot(1, len(images), i + 1, xticks = [], yticks = [])
        image = image.numpy().transpose((1, 2, 0))
        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229, 0.224, 0.225])
        image = image*std + mean
        image = np.clip(image, 0., 1.)
        plt.imshow(image)
        col = 'green'
        if preds[i] != labels[i]:
            col = 'red'
            
        plt.xlabel(f'{class_names[int(labels[i].numpy())]}')
        plt.ylabel(f'{class_names[int(preds[i].numpy())]}', color = col)
    plt.tight_layout()
    plt.show()
```

# Utils

```python
class AverageMeter(object):
    """Computes and stores the average and current value"""
    def __init__(self):
        self.reset()

    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count
```

# Train Model

```python
def train_model(epochs: int):
    print('Starting training...')
    for epoch in range(epochs):
        loss_train = AverageMeter()
        acc_train = Accuracy(task='multiclass', num_classes=3).to(device)

        model.train() # set model to training phase

        for train_step, (images, labels) in enumerate(train_loader):
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = loss_fn(outputs, labels)

            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

            loss_train.update(loss.item())
            acc_train(outputs, labels.int())

        print(f'Epoch: {epoch}')
        print(f'Train: Loss={loss_train.avg:.2f}, Accuracy={100. * acc_train.compute().item()}')
        print()
```

```python
def train_model(model, train_loader, loss_fn, optimizer):
    start = time.time()
    model.train()
    loss_train = AverageMeter()
    acc_train = Accuracy(task='multiclass', num_classes=3).to(device)
    for i, (inputs, targets) in enumerate(train_loader):
        inputs = inputs.to(device)
        targets = targets.to(device)       
        
        outputs = model(inputs)
        loss = loss_fn(outputs, targets)
        
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
        loss_train.update(loss.item())
        acc_train(outputs, targets) 
        
    finish = time.time() - start   
    for _ in track(range(train_loader.batch_size), description='[green]Training...'):
        time.sleep(finish)
        
    return model, loss_train.avg, acc_train.compute()
```

```python
train_model(3)
```

# Evaluation Model

```python
def eval_model():
  with torch.no_grad():
    loss_eval = AverageMeter()
    acc_eval = Accuracy(task='multiclass', num_classes=3).to(device)

    for i, (inputs, targets) in enumerate(test_loader):
      inputs = inputs.to(device)
      targets = targets.to(device)

      outputs = model(inputs)
      loss = loss_fn(outputs, targets)
      
      loss_eval.update(loss.item())
      acc_eval(outputs, targets.int())

    print(f'Evaluation: Loss={loss_eval.avg:.2f}, Accuracy={100. * acc_eval.compute().item():.2f}')
    show_preds()
```

```python
# evaluate function
def evaluate(model, valid_loader, loss_fn):
    model.eval()
    with torch.no_grad():
        loss_valid = AverageMeter()
        acc_valid = Accuracy(task='multiclass', num_classes=4).to(device)
        
        for i, (inputs, targets) in enumerate(valid_loader):
            inputs = inputs.to(device)
            targets = targets.to(device) 
                
            outputs = model(inputs)
            loss = loss_fn(outputs, targets)
                
            loss_valid.update(loss.item())
            acc_valid(outputs, targets)
            
    return loss_valid.avg, acc_valid.compute()
```
